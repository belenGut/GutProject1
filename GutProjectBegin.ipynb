# Analysis #
Analysis is difficult because the data in both spreadsheets is put together in a confusing way. For questions that let students choose multiple answers, each individual option is given its own column, with the data being either 'Yes' or 'No'. This does prevent cells from having multiple answers (i.e. a single cell that says 'social media', 'community event'). The columns are still in conversational English, which means the titles are long and hard to be all displayed.

Something went wrong with the column asking if students found out about CCM from the CCM website. It is placed before all the other columns, including the timestamp column. I believe this is why it is included in subsets that should not have it.

I decided to look at whether students found out about CCM from a high school teacher. The corresponding response in the 'CCM Web site column' is included for reasons stated above. In both datasets, more people learned about CCM from a teacher than did not, but the difference is not very large. I made two big subsets which I then split even further. The subset 'comTeach' contains all responses to the high school teacher question in the first dataset. The subset 'litTeach' contains all responses to the high school teacher question in the second dataset. For both of those subets, I created two more subsets, one containing all the 'Yes' responses and the other containing all the 'No' responses. Comparing the lengths of 'comTeach' and 'litTeach' with their subsets reveal that all responses were either 'Yes', 'No', or 'Don't recall', with no null values. Seeing that the each main subset is as long as their parent subset confirms this.
